****PREDICTING MOLAR SOLUBILITY WITH LINEAR REGRESSION****


**John S. Delaney(1)** makalesinde moleküllerin 4 fizikokimyasal özelliği;

*   cLogP (Octanol-water partition coefficient)
*   MW (Molecular weight)
*   RB (Number of rotatable bonds)
*   AP (Aromatic proportion = number of aromatic atoms / total number of heavy atoms)

baz alınarak linear regression modeli ile çözünürlük tahmin algoritması yazmanın mümkün olduğu ve üretilen herhangi bir ilacın sudaki çözünürlüğünü tahmin etmenin mümkün olduğunu açıklamaktadır.

```python
import numpy as np
from rdkit.Chem import Descriptors
```

```python
def generate(smiles, verbose=False):

    moldata= []
    for elem in smiles:
        mol=Chem.MolFromSmiles(elem) 
        moldata.append(mol)
       
    baseData= np.arange(1,1)
    i=0  
    for mol in moldata:        
       
        desc_MolLogP = Descriptors.MolLogP(mol)
        desc_MolWt = Descriptors.MolWt(mol)
        desc_NumRotatableBonds = Descriptors.NumRotatableBonds(mol)
           
        row = np.array([desc_MolLogP,
                        desc_MolWt,
                        desc_NumRotatableBonds])   
    
        if(i==0):
            baseData=row
        else:
            baseData=np.vstack([baseData, row])
        i=i+1      
    
    columnNames=["MolLogP","MolWt","NumRotatableBonds"]   
    descriptors = pd.DataFrame(data=baseData,columns=columnNames)
    
    return descriptors
```

```python
df = generate(data.SMILES)
df
     
```

```python
def AromaticAtoms(m):
  aromatic_atoms = [m.GetAtomWithIdx(i).GetIsAromatic() for i in range(m.GetNumAtoms())]
  aa_count = []
  for i in aromatic_atoms:
    if i==True:
      aa_count.append(1)
  sum_aa_count = sum(aa_count)
  return sum_aa_count
```

```python
X = pd.concat([df,df_desc_AromaticProportion], axis=1)
X
```

```python
data.head()
```

```python

Y = data.iloc[:,1]
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
     
```

```python
Y_pred_train = model.predict(X_train)
     
```

```python
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(Y_train, Y_pred_train))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(Y_train, Y_pred_train))
```

```python
Y_pred_test = model.predict(X_test)
```

```python
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(Y_test, Y_pred_test))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(Y_test, Y_pred_test))
```

The work of Delaney
 provided the following linear regression equation:

LogS = 0.16 - 0.63 cLogP - 0.0062 MW + 0.066 RB - 0.74 AP

Based on the Train set
LogS = 0.30 -0.75 LogP - .0066 MW -0.0041 RB - 0.36 AP

Based on the Full dataset
LogS = 0.26 -0.74 LogP - 0.0066 + MW 0.0032 RB - 0.42 AP

```python
print('LogS = %.2f %.2f LogP %.4f MW %.4f RB %.2f AP' % (model.intercept_, model.coef_[0], model.coef_[1], model.coef_[2], model.coef_[3] ) )
```
